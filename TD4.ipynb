```python
import pandas as pd

# Charger la carte à partir du fichier texte
map = pd.read_csv('monlaby_3.txt', delimiter=' ', header=None)
map = pd.DataFrame(map)

# Initialiser l'aventurier
aventurier = {}
aventurier["position"] = [6, 0]
aventurier["etat"] = 0

# Trouver la position du joyau sur la carte
joyau_position = None
for i in range(map.shape[0]):
    for j in range(map.shape[1]):
        if map.iloc[i,j] == '*':
            joyau_position = [i, j]
            break
    if joyau_position is not None:
        break
aventurier["joyau_position"] = joyau_position

# Définir les mouvements possibles
moves = [(1, 0), (-1, 0), (0, 1), (0, -1)]

# Fonction pour obtenir les positions possibles à partir d'une position actuelle et d'un mouvement
def get_positions_possibles(x, m):
    return (x[0]+m[0], x[1]+m[1])

# Fonction pour vérifier si une position est autorisée sur la carte
def est_autorise(position, carte):
    x , y  = position
    if x < 0 or x >= carte.shape[0] or y < 0 or y >=  carte.shape[1]:
        return False
    if carte.iloc[x,y] == 1:
        return False
    return True

# Initialiser la liste des positions à explorer et des positions déjà explorées
a_explorer = [aventurier['position']]
deja_explorer = [aventurier['position']]

# Boucle principale pour explorer la carte
while len(a_explorer) > 0:
    print(map)
    action = input("Appuyez sur Entrée pour continuer ou Q pour quitter : ")
    if action.lower() == 'q':
        break
    print(aventurier["joyau_position"])    
    if aventurier["joyau_position"] == aventurier["position"]:
        print('Vous avez trouvé le joyau !')
        break
    else:
        print('Position actuelle :', aventurier["position"])
    for out in moves:
        nouvelle_position = get_positions_possibles(aventurier["position"], out)
        if est_autorise(nouvelle_position, map) and nouvelle_position not in deja_explorer:
            a_explorer.append(nouvelle_position)
            deja_explorer.append(nouvelle_position)
            # Mettre à jour la carte en marquant la nouvelle position
            map.iloc[nouvelle_position[0], nouvelle_position[1]] = 'A'
    # Choisir la prochaine position à explorer en prenant la première de la liste
    aventurier['position'] = a_explorer.pop(0)


```
