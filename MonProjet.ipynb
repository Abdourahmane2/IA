```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder


def transform(df):   #transform   les boollens vrai et faux en 1 et 0 
    for c in df.columns:
        if df[c].dtype=='object':
            df[c] = df[c].fillna('N')
            lbl = LabelEncoder()
            lbl.fit(list(df[c].values))
            df[c] = lbl.transform(df[c].values)
    return df

# lire le fichier CSV
database0 = pd.read_csv('zoo.csv')

database0['animal'] = database0['animal'].apply(lambda x : x.replace('b' , ''))

# appliquer la transformation des données
database = transform(database0.copy())

# supprimer les colonnes inutiles
database.drop(['airborne', 'fins', 'legs', 'catsize'], axis=1, inplace=True)



# séparer les données en entrainement et test
dataY = database['type']
dataX = database.drop(['animal', 'type'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(dataX, dataY, test_size=0.2, random_state=42)

# entrainer le modèle
trun = DecisionTreeClassifier()
trun.fit(X_train, y_train)

data3=database.copy()
data3['type']=database['type']
data3['animal']=database['animal']

# demander les caractéristiques de l'animal
hair = input('A-t-il des cheveux? (oui/non) : ').lower() == 'oui'
fea = input('A-t-il des plumes? (oui/non) : ').lower() == 'oui'
egg = input('Pond-il des oeufs? (oui/non) : ').lower() == 'oui'
milk = input('Boit-il du lait? (oui/non) : ').lower() == 'oui'
aqua = input('Est-il aquatique? (oui/non) : ').lower() == 'oui'
pred = input('Est-il un prédateur? (oui/non) : ').lower() == 'oui'
dent = input('A-t-il des dents? (oui/non) : ').lower() == 'oui'
colo = input('A-t-il une colonne vertébrale? (oui/non) : ').lower() == 'oui'
brea = input('Respire-t-il par les poumons? (oui/non) : ').lower() == 'oui'
ven = input('Est-il venimeux? (oui/non) : ').lower() == 'oui'
que = input('A-t-il une queue? (oui/non) : ').lower() == 'oui'
domes = input('Est-il domestique? (oui/non) : ').lower() == 'oui'

# prédire l'animal
resultats = pd.DataFrame([[int(hair), int(fea), int(egg), int(milk), int(aqua), int(pred), int(dent), int(colo), int(brea),
              int(ven), int(que), int(domes)]], columns=dataX.columns)

# Predict the animal
predicted_class = trun.predict(resultats)
animal_name = database0.loc[data3['type'] == predicted_class[0], 'animal'].iloc[0]
print("I think you're thinking of:", animal_name)

```
